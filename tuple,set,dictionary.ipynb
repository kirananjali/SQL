TUPLE
Tuple items are ordered, unchangeable, and allow duplicate values.

```python
thistuple = ("apple", "banana", "cherry")
print(len(thistuple))
```

To create a tuple with only one item, you have to add a comma after the item, otherwise Python will not recognize it as a tuple.

```python
thistuple = ("apple",)
print(type(thistuple))

#NOT a tuple
thistuple = ("apple")
print(type(thistuple))
```

```python
#A tuple can contain different data types:
tuple1 = ("abc", 34, True, 40, "male")
```

```python
#access tuple items by referring to the index number 
thistuple = ("apple", "banana", "cherry")
print(thistuple[1])
```

```python
#Negative indexing means start from the end.
thistuple = ("apple", "banana", "cherry")
print(thistuple[-1])
```

Range of Indexes


```python
thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[2:5])
print(thistuple[:4])
print(thistuple[2:])
print(thistuple[-4:-1]) #negative indexing
```

```python
if "apple" in thistuple:
  print("Yes, 'apple' is in the fruits tuple")
```

```python
#You can convert the tuple into a list, change the list, and convert the list back into a tuple.
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)

print(x)
```

```python
y.append("sapota")
x = tuple(y)

print(x)
```

```python
y.remove("kiwi")
x = tuple(y)

print(x)
```

```python
# METHODS
x = (1,5,3,3,7,8,4,9,9)
count() -> x.count(3)
index() -> x.index(9) #Search for the first occurrence of the value 8
```

SETS
A set is a collection which is unordered, unchangeable*, and unindexed.

Once a set is created, you cannot change its items, but you can remove items and add new items.

```python
# True and 1 is considered the same value, False and 0 is considered the same value:

thisset = {"apple", "banana", "cherry", True, 1, 2}

print(thisset)
```

```python
#A set with strings, integers and boolean values:

set1 = {"abc", 34, True, 40, "male"}
```

```python
#Access Items
#You cannot access items in a set by referring to an index or a key.
#But you can loop through the set items using a for loop, or ask if a specified value is present in a set, by using the in keyword.
```

```python
thisset = {"apple", "banana", "cherry"}

for x in thisset:
  print(x)
```

```python
thisset = {"apple", "banana", "cherry"}

print("banana" in thisset)
```

```python
thisset.add("sapota") #add items
print(thisset)
```

```python
set2 = {"mango","opranges"}
thisset.update(set2) #add sets
print(thisset)
```

```python
lst=["eggs",1,55]
thisset.update(lst)
print(thisset)
```

```python
#If the item to remove does not exist, remove() will raise an error.
#If the item to remove does not exist, discard() will NOT raise an error.
thisset.remove("eggs")
print(thisset)
```

```python
#Sets are unordered, so when using the pop() method, you do not know which item that gets removed.
x = thisset.pop()

print(x)

print(thisset)
```

DICTONARIES
Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates.

```python
d1 = {"name": "kiran",
      "age": 23,
    "loc": "USA"}
print(d1["age"]) #can be referred to by using the key name.
```

```python
d1 = {"name": "kiran",
      "age": 23,
    "loc": "USA",
     "loc": "INDIA"} #Dictionaries cannot have two items with the same key
print(d1["loc"])
```

```python
print(len(d1))
```

```python
x = d1.get("name") #can access item by referring to it's key name, by using get() method.
print(x)
```

```python
#get keys
x = d1.keys()
print(x)
```

```python
x = d1.values()
print(x)
```

```python
x = d1.items()
print(x)
```

```python
#add new item
d1["DL"] = "yes"
print(d1)
```

```python
#update dictionary
d1.update({'loc':'dubai'})
print(d1)
```

```python
#pop
#popitem() method -> removes the last inserted item( but in 3.7 version, randomly removes item instead)
d1.pop('DL')
print(d1)
```
