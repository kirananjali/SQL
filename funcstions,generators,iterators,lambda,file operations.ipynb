From a function's perspective:

A parameter is the variable listed inside the parentheses in the function definition.

An argument is the actual value that is sent to the function when it is called.

```python
def my_func(name): # name is parameter
    print("hello", name)
    
my_func("kiran") #"kiran" is argumnet
```

By default, a function must be called with the correct number of arguments.
*args and **kwargs allow functions to accept a unknown number of arguments.

```python
#This way, the function will receive a tuple of arguments and can access the items accordingly
def my_func(*fruits):
    print("shop1:", fruits[0])
    print("shop3:", fruits[2])
    print("all shops:", fruits)
my_func("orange","banana","apple","sapota")
```

```python
def my_func(owner, *shops):
    for i in shops:
        print(owner, i)
        
my_func("Pruthvi","grocery", "restaurant")
```

#'**kwargs'
#the function will receive a dictionary of arguments and can access the items accordingly

```python
def my_func(**student):
    print("My first name is", student["fname"], "and my last name is", student["lname"])
    
my_func(fname = "Kirananjali", lname = "Bethu")
```

```python
#Regular parameters must come before *args,**kwargs:
def my_func(title, *args, **kwargs):
    print("Title:", title)
    print("Positional argument:", args)
    print("keyword argument:", kwargs)
    
my_func('Student info','kiran','rushi', age=23, state = 'Florida')
```

#**Recursion
Recursion is when a function calls itself.
This has the benefit of meaning that you can loop through data to reach a result.

```python
# factorial of a number
def factorial(n):
    #base case - A condition that stops the recursion
    if n==0 or n==1:
        return 1
    #recursive case - The function calling itself with a modified argument
    else:
        return n * factorial(n-1)
    
print(factorial(8)) # 8*7*6*5*4*3*2*1
```

The Fibonacci sequence is a classic example where each number is the sum of the two preceding ones. The sequence starts with 0 and 1:

0, 1, 1, 2, 3, 5, 8, 13, ...

We can use recursion to find a specific number in the sequence

```python
def fibonacci(n):
    if n<=1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
print(fibonacci(6))
```

#generator
#special type of iterator uses 'yield' instead of return

Generators allow you to iterate over data without storing the entire dataset in memory.

Generators are memory-efficient because they generate values on-the-fly instead of storing everything in memory.
Instead of using return, generators use the yield keyword.

#yield
When yield is encountered, the function's state is saved, and the value is returned. The next time the generator is called, it continues from where it left off.

```python
def large_sequence(n):
  for i in range(n):
    yield i

# This doesn't create a million numbers in memory
gen = large_sequence(1000000)
print(next(gen))
print(next(gen))
print(next(gen))
```

LAMBDA, (anonymous function.)
IT'S LIKE TEMPORARY FUNCTION
Lambda functions can take any number of arguments
SYNTAX:
lambda arguments : expression

```python
x = lambda a,b : a*b
print(x(2,4))
```

```python
x = lambda a, b, c : a * b + c
print(x(5, 6, 2))
```

WHY LAMBDA
The power of lambda is better shown when you use them as an anonymous function inside another function.

```python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)
mytripler = myfunc(3)

print(mydoubler(11))
print(mytripler(11))
```

Lambda with Built-in Functions
Lambda functions are commonly used with built-in functions like map(), filter(), and reduce()

```python
#Syntax
#map(function, iterable,..)

def double(val):
    return val * 2

a = [1, 2, 3, 4]    
res = list(map(double, a))
print(res)
```

```python
res = list(map(lambda x: x**2,a))
print(res)
```

```python
def summation(a):
    return a+2
def mul(a):
    return a**2
a = [2,4,6,8]
cal = list(map(summation,a))
cal1 = list(map(mul,a))

print(cal)
print(cal1)
```

```python
#filter
#The filter() function creates a list of items for which a function returns True.
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

```python
#reduce
from functools import reduce
max_find = lambda a,b: a if a>b else b
lst = [5,77,33,8,200]

reduce(max_find,lst)
```

"t" - Text - Default value. Text mode
